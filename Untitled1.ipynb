```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt

import pandas as pd 
import numpy as np
```

```python
file1 = pd.read_csv('model已标注.csv', encoding='gbk')
file2 = pd.read_csv('model1已标注.csv', encoding='gbk')

file = pd.concat([file1,file2])
```

```python
file_para = file.loc[file['type']=='para']
```

```python
pdf_page_dict = {
    '页眉':0,
    '页脚':1,
    '目录':2,
    '封面':2,
    '正文':2,
    '标题1':2,
    '标题2':2,
    '标题3':2,
    '标题4':2,
    '标题5':2,
    '标题6':2,
    '其它':2
}
```

```python
file_para['Unnamed: 21'] = file_para['Unnamed: 21'].apply(lambda x: pdf_page_dict[x])
```

```python
file_para.head(5)
```

```python
file_para.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
file_para.drop(['object_type','fontname','type','upright'],axis=1,inplace=True)
```

```python
file_para.head(5)
```

```python
file_para['len_str'] = file_para['text'].apply(lambda x: len(x))
```

```python
file_para = file_para.rename(columns={'Unnamed: 21': 'page_type'})
```

```python
file_para.head(20)
```

```python
file_para['is_mulu'] = file_para['text'].apply(lambda x:1 if '......' in x else 0)
```

```python
import re
def get_type(text):
    if re.search('^[\(（]?[一二三四五六七八九十]+[\)）]?',text):
        return 1
    else:
        return 0

file_para['is_biaoti'] = file_para['text'].apply(get_type)    
```

```python
file_para
```

```python
y_file = file_para['page_type']
x_file = file_para.drop('page_type',axis=1)
```

```python
x_file = x_file.drop('text', axis=1)
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x_file, y_file, test_size=0.3)

clf = RandomForestClassifier()
clf.fit(Xtrain, Ytrain)

score = clf.score(Xtest, Ytest)
score
```

```python
clf = RandomForestClassifier(n_estimators=100)

clf = cross_val_score(clf, x_file, y_file, cv=10)

clf
```

```python
score_list = []

for i in range(50):
    clf = RandomForestClassifier(n_estimators=100,max_depth=1+i,bootstrap=True)
    clf = cross_val_score(clf, x_file, y_file, cv=10)
    score_list.append(clf.mean())
    
print(max(score_list))
plt.plot(range(1,51),score_list)
plt.show()
```
